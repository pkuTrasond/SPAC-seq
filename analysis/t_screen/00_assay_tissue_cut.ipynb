```python
from skimage import io
from skimage.color import rgb2gray

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# see stereo-seq chip coordinate
chip_df = pd.read_csv('./RNA.raw.combine.gem', sep='\t',comment = '#')
```

```python
# aggregate chip coordinate by RNA
chip_df = chip_df.groupby(['x', 'y']).sum().reset_index()
w = chip_df['x'].max() + 1
h = chip_df['y'].max() + 1
chip = np.zeros((w, h))
chip[chip_df['x'], chip_df['y']] = chip_df['MIDCount']
chip *= 25
chip = np.clip(chip, 0, 255)
```

```python
# plot chip RNA intensity
plt.imshow(chip, cmap='gray')
```
